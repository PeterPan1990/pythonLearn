```python
# cookie的应用，模拟登陆
import urllib2
import cookielib

cookie = cookielib.CookieJar() # 声明一个CookieJar对象实例来保存cookie
handle = urllib2.HTTPCookieProcessor(cookie) # 利用urllib2库中HTTPCookieProcessor 对象来创建cookie处理器
opener = urllib2.build_opener(handle) # 用handler来构建opener

response = opener.open('http://www.baidu.com') # open方法同urllib2中的urlopen方法，也可以传入request
for item in cookie:
    print 'Name = ' + item.name
    print 'Value = ' + item.value
    
#结果：
#Name = BAIDUID
#Value = 46AB1386DF1DB6D10AAA9C1743D5BCA9:FG=1
#Name = BIDUPSID
#Value = 46AB1386DF1DB6D10AAA9C1743D5BCA9
#Name = H_PS_PSSID
#Value = 10299_13424_1431_14355_13518_13075_12867_14167_14295_10563_12722_14155_14173_13203_14244_11911_13936_14268_14183_8498_10634_14195
#Name = BDSVRTM
#Value = 0
#Name = BD_HOME
#Value = 0
```

```python
# 将cookie保存到文件中
import cookielib
import urllib2

filename = 'cookie.txt' # 设置文件名
cookie = cookielib.MozillaCookieJar(filename) # 申明一个MozillaCookieJar对象实例来保存cookie，然后写入文件
handle = urllib2.HTTPCookieProcessor(cookie)
opener = urllib2.build_opener(handle)
response = opener.open('http://www.baidu.com')
cookie.save(ignore_discard=True, ignore_expires=True)
```

```python
# 从cookie文件中获取cookie并访问
import cookielib
import urllib2

cookie = cookielib.MozillaCookieJar() # 创建MozillaCookieJar对象
cookie.load('cookie.txt', ignore_discard=True, ignore_expires=True) # 从文件读取cookie内容到变量
req = urllib2.Request('http://www.baidu.com')
opener = urllib2.build_opener(urllib2.HTTPCookieProcessor(cookie)) # 创建一个opener
#print response.read()
```

```python
# 正则表达式
import re

pattern = re.compile(r'hello') #将正则表达式编译为pattern对象

result1 = re.match(pattern, 'hello') # 使用re.match匹配文本
result2 = re.match(pattern, 'helloo CQC!')
result3 = re.match(pattern, 'helo CQC!')
result4 = re.match(pattern, 'hello CQC!')

if result1:
    print result1.group() # 使用match获得分组信息
else:
    print '1匹配失败'
    

if result2:
    print result2.group()
else:
    print '2匹配失败'
    

if result3:
    print result3.group()
else:
    print '3匹配失败'
    

if result4:
    print result4.group()
else:
    print '4匹配失败'
```
